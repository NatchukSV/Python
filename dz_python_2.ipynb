```python
### Task 2
sum_n = 0
while sum_n > -1:
      n = int(input('Введите число'))
      sum_n += n

      if n == 0:
          print('Результат:', sum_n)
          break
```

```python
### Task 3
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
boys_s = sorted(boys)
girls_s = sorted(girls)
par = []
sorta = 0

if len(boys_s) == len(girls_s):
    while sorta < len(boys_s):
        para = boys_s[sorta] + ' и ' + girls_s[sorta]
        par.append(para)
        sorta += 1 
else:
    print('Внимание, кто-то может остаться без пары!')
par
```

```python
### Task 4
countries_temperature = [
                        ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
                        ['Germany', [57.2, 55.4, 59, 59, 53.6]],
                        ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
                        ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
                        ]
x = 0
print('Средняя температура в странах:')
while x < len(countries_temperature):
    print(countries_temperature[x][0], '-', round(((sum(countries_temperature[x][1]) / len(countries_temperature[x][1]) - 32) * 5 / 9), 1), 'C')
    x += 1 
```

```python
### Task 6
data = input('Введите набор чисел, разделенные пробелом').split()
result = {}

for namber in set(data):
    result[namber] = data.count(namber)

for del_res in set(result):
     if result[del_res] == 1:
         result.pop(del_res)
         
print(list(result.keys()))
```
