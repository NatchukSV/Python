```python
### Task 5 # Если не использовать списки + сплит конвертацию значений
number = 153351
sum_start = 0
sum_end = 0
i = 1 
for i in 0, 1, 2:
    sum_start += (number // 10**(i + 3)) % 10
    sum_end += (number // 10**i) % 10
    i += 1
if sum_start == sum_end:
    print ('Счастливый билет')
else: 
    print ('Несчастливый билет')
```

```python
### Task 6
figure = str(input('Введите фигуру')) 
if figure == 'Круг':
    r = int(input('Введите радиус круга (см.)'))
    print ('Площадь Круга','%.2f' % 3.14*r**2)
elif figure == 'Треугольник':
    a = int(input('Введите длину стороны A (см.)'))
    b = int(input('Введите длину стороны B (см.)'))
    c = int(input('Введите длину стороны C (см.)'))
    p = (a + b + c) / 2
    S = (p * (p - a) * (p - b) * (p - c))**(1./2)
    print ('Площадь Треугольника',"%.2f" % S)
elif figure == 'Прямоугольник':
    a = int(input('Введите длину стороны A (см.)'))
    b = int(input('Введите длину стороны B (см.)'))
    S = a * b
    print ('Площадь Прямоугольника',"%.2f" % S)
else:
    print ('Вы неверно указали фигуру')
```
